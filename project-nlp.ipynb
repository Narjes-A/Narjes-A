```python
def process_video_slice(video_files, start_idx, end_idx, model):
    """
    Process a slice of video files from start_idx (inclusive) to end_idx (exclusive).
    Downloads each video, extracts audio, transcribes it with Whisper,
    saves the transcript locally, and uploads it to your S3 bucket.
    """
    for video_key in video_files[start_idx:end_idx]:
        # Determine local file path for the video
        local_video_file = os.path.join('videos', os.path.basename(video_key))
        print(f"\nDownloading {video_key} to {local_video_file} ...")
        s3.download_file(source_bucket, video_key, local_video_file)
        
        # Extract audio from the video and save as WAV
        local_audio_file = os.path.join('audio', os.path.basename(video_key).replace('.mp4', '.wav'))
        print(f"Extracting audio from {local_video_file} to {local_audio_file} ...")
        try:
            clip = mp.VideoFileClip(local_video_file)
            clip.audio.write_audiofile(local_audio_file, logger=None)
        except Exception as e:
            print(f"Error extracting audio from {local_video_file}: {e}")
            continue
        
        # Transcribe audio using Whisper
        print(f"Transcribing audio from {local_audio_file} ...")
        try:
            result = model.transcribe(local_audio_file)
            transcript_text = result['text']
        except Exception as e:
            print(f"Error transcribing {local_audio_file}: {e}")
            continue
        
        # Save transcript to a text file
        transcript_file = os.path.join('transcripts', os.path.basename(video_key).replace('.mp4', '.txt'))
        with open(transcript_file, 'w', encoding='utf-8') as f:
            f.write(transcript_text)
        print(f"Transcript saved to {transcript_file}")
        
        # Upload the transcript to your personal S3 bucket
        s3.upload_file(transcript_file, output_bucket, f"transcripts/{os.path.basename(transcript_file)}")
        print(f"Transcript uploaded to s3://{output_bucket}/transcripts/{os.path.basename(transcript_file)}")
        
        print("-" * 50)
```

```python
# Define a function to normalize text
def normalize_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and numbers (only keep letters and whitespace)
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize the text into words
    words = word_tokenize(text)
    # Remove English stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    # Return the normalized text as a single string
    return " ".join(words)

# S3 setup: Define my bucket and the folder where transcripts are stored
bucket_name = 'narjes-project-bucket'
transcript_prefix = 'transcripts/'  # Define folder path where transcripts are stored.

# Initialize the boto3 S3 client
s3 = boto3.client('s3')

# Create a local directory to store normalized transcripts
os.makedirs('normalized_transcripts', exist_ok=True)

# List all files in the S3 bucket under the "transcripts/" folder
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=transcript_prefix)

# Print a message if no transcripts found.
if 'Contents' not in response:
    print("No transcripts found in the bucket!")
else:
    for obj in response['Contents']:
        key = obj['Key']
        if key.endswith('.txt'):  #Process only .txt files, skipping any other files.
            # Extract the filename, and download the file from S3.
            local_file = os.path.join('transcripts', os.path.basename(key))
            print(f"Downloading {key} as {local_file}...")
            s3.download_file(bucket_name, key, local_file)
            
            # Open and read the transcript from the downloaded file.
            with open(local_file, 'r', encoding='utf-8') as f:
                transcript_text = f.read()
            
            # Normalize the transcript text
            normalized_text = normalize_text(transcript_text)
            
            # Save the normalized text into a new file
            normalized_filename = os.path.join('normalized_transcripts', f"normalized_{os.path.basename(key)}")
            with open(normalized_filename, 'w', encoding='utf-8') as f:
                f.write(normalized_text)
                
            print(f"Normalized transcript saved to {normalized_filename}")
```

```python
'''
Downloads normalized transcript files (.txt) from an S3 bucket.
Extracts key phrases from the transcripts using TF-IDF (important words and phrases).
Performs topic modeling using LDA to find main topics in the text.

'''


# 1. Download Normalized Transcripts from S3
# ---------------------------
bucket_name = 'narjes-project-bucket'
normalized_prefix = 'normalized_transcripts/'

s3 = boto3.client('s3')

# List objects(files) in the normalized transcripts folder of your bucket
response = s3.list_objects_v2(Bucket=bucket_name, Prefix=normalized_prefix)
documents = []

# Check if there are any files in the response
if 'Contents' in response:
    for obj in response['Contents']:
        key = obj['Key']  # Get the file path (key) in the S3 bucket
        
        # Only process .txt files (ignores other file types)
        if key.endswith('.txt'):
            local_filename = os.path.basename(key)   # Extract the filename from the full S3 path
            print(f"Downloading {key} as {local_filename}...")
            s3.download_file(bucket_name, key, local_filename)            # Download the file from S3 to the local directory

            # Open and read the contents of the file, then store it in the `documents` list
            with open(local_filename, 'r', encoding='utf-8') as f:
                documents.append(f.read())
            os.remove(local_filename)             # Clean up the local file after reading its content

else:
    print("No normalized transcripts found in the bucket!")

print(f"Number of documents loaded: {len(documents)}")

# ---------------------------
# 2. Extracting Key Phrases using TF-IDF
# ---------------------------

"""
    Extracts key phrases from a list of text documents using TF-IDF.
    :texts: List of strings (documents)
    :num_keywords: Number of top keywords/phrases to extract per document
    :return: List of key phrases for each document
"""
def extract_key_phrases(texts, num_keywords=3):
    # Create a TF-IDF vectorizer that considers unigrams and bigrams
    vectorizer = TfidfVectorizer(ngram_range=(1, 2), stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(texts)      # Transform the text into a TF-IDF matrix    
    feature_names = vectorizer.get_feature_names_out()      # Get the list of words/phrases from the vectorizer
    dense = tfidf_matrix.todense().tolist()       # Convert the sparse TF-IDF matrix to a dense list format for easier processing

    
    keywords = []
    # For each document, pick the top 'num_keywords' features based on TF-IDF scores
    for doc in dense:
        top_indices = sorted(range(len(doc)), key=lambda i: doc[i], reverse=True)[:num_keywords]
        keywords.append([feature_names[i] for i in top_indices])          # Retrieve the corresponding words/phrases from the feature names

    
    return keywords

# Example extraction of key phrases from the documents
key_phrases = extract_key_phrases(documents, num_keywords=3)
print("\nKey Phrases for each document:")
for i, kp in enumerate(key_phrases):
    print(f"Document {i+1}: {kp}")

# ---------------------------
# 3. Topic Modeling using LDA
# ---------------------------

"""
    Extracts topics from a list of text documents using Latent Dirichlet Allocation (LDA).
    :texts: List of strings (documents)
    :num_topics: Number of topics to extract
    :num_words: Number of top words per topic
    :return: List of topics (each topic is a list of words)
"""

def extract_topics(texts, num_topics=2, num_words=3):
    # Create a document-term matrix using CountVectorizer
    vectorizer = CountVectorizer(stop_words='english')   # Remove common English stopwords
    doc_term_matrix = vectorizer.fit_transform(texts)    # Convert text to a matrix format

    # Initialize and fit LDA on the document-term matrix
    lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
    lda.fit(doc_term_matrix)

    words = vectorizer.get_feature_names_out()      # Get the list of words from the vectorizer
    topics = []
    # For each topic, select the top 'num_words' words
    for topic_idx, topic in enumerate(lda.components_):
        top_words = [words[i] for i in topic.argsort()[-num_words:]]
        topics.append(top_words)

    return topics

# Example extraction of topics from the corpus
topics = extract_topics(documents, num_topics=2, num_words=3)
print("\nTopics extracted from the corpus:")
for idx, topic in enumerate(topics):
    print(f"Topic {idx+1}: {topic}")
```

```python
def generate_thumbnail(video_path, thumbnail_path, t=1.0):
    """
    Extract a frame from the given video_path at time t (seconds)
    and save it as an image at thumbnail_path.
    """
    try:
        clip = mpe.VideoFileClip(video_path)          # Load the video file into a VideoFileClip object
        clip.save_frame(thumbnail_path, t=t)          # Extract and save a frame from the video at time t (default: 1.0 sec)
        clip.close()        # Close the video file to free resources

    except Exception as e:
        print(f"Error generating thumbnail for {video_path}: {e}")
```

```python
# 1. Set Up Flask Server
# -------------------------------
server = flask.Flask(__name__)

# -------------------------------
# 2. Data Loading and Helper Functions
# -------------------------------
def get_thumbnail(video_title):
    """Return the path or URL of a thumbnail image for a given video title."""
    thumb_path = os.path.join("thumbnails", video_title + ".jpg")
    if os.path.exists(thumb_path):
        return "/" + thumb_path
    else:
        return "https://via.placeholder.com/300x200?text=No+Thumbnail"

def extract_key_phrases(texts, num_keywords=3):
    """Use TF-IDF to extract top key phrases from a list of documents."""
    vectorizer = TfidfVectorizer(ngram_range=(1, 2), stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(texts)
    feature_names = vectorizer.get_feature_names_out()
    dense = tfidf_matrix.todense().tolist()
    key_phrases = []
    for doc in dense:
        if sum(doc) == 0:
            key_phrases.append("")
        else:
            top_indices = sorted(range(len(doc)), key=lambda i: doc[i], reverse=True)[:num_keywords]
            key_phrases.append(", ".join([feature_names[i] for i in top_indices]))
    return key_phrases

# -------------------------------
# 3. Build the DataFrame
# -------------------------------
normalized_folder = 'normalized_transcripts'
file_names, documents = [], []

for filename in os.listdir(normalized_folder):
    if filename.endswith('.txt'):
        file_path = os.path.join(normalized_folder, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        file_names.append(filename)
        documents.append(content)

video_titles = [fname.replace('normalized_', '').replace('.txt', '') for fname in file_names]
df = pd.DataFrame({
    'video_title': video_titles,
    'normalized_text': documents
})

df['key_phrases'] = extract_key_phrases(df['normalized_text'])

# Adding video and thumbnail columns
df['video_file'] = "videos/" + df['video_title'] + ".mp4"
df['thumbnail'] = df['video_title'].apply(get_thumbnail)

# -------------------------------
# 3.1 Topic Modeling to Extract Topic Words
# -------------------------------
# We'll use CountVectorizer and LDA to extract topics.
num_topics = 2    # The number of topics.
num_words = 3     # Number of top words per topic.

count_vectorizer = CountVectorizer(stop_words='english')
doc_term_matrix = count_vectorizer.fit_transform(df['normalized_text'])

lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(doc_term_matrix)

# Get the top words for each topic.
feature_names = count_vectorizer.get_feature_names_out()
topic_words = []
for topic in lda.components_:
    top_indices = topic.argsort()[-num_words:]
    top_words = [feature_names[i] for i in top_indices]
    topic_words.append(", ".join(top_words))

# For each document, determine the dominant topic and store its top words.
topic_assignments = lda.transform(doc_term_matrix)  # shape: (n_docs, num_topics)
dominant_topics = topic_assignments.argmax(axis=1)
df['topics'] = [topic_words[i] for i in dominant_topics]


# -------------------------------
# 4. Optional: Serve Local Videos/Images
# -------------------------------
@server.route('/videos/<path:filename>')
def serve_video(filename):
    return flask.send_from_directory('videos', filename)

@server.route('/thumbnails/<path:filename>')
def serve_thumbnail(filename):
    return flask.send_from_directory('thumbnails', filename)


# -------------------------------
# 5. Initialize the Dash App
# -------------------------------
app = JupyterDash(__name__, server=server, external_stylesheets=[dbc.themes.BOOTSTRAP])

# -------------------------------
# 6. Layout
# -------------------------------
app.layout = dbc.Container([
    dbc.Row([
        dbc.Col(
            dcc.Input(
                id='search-input',
                type='text',
                placeholder='Enter keyword...',
                style={'width': '100%', 'padding': '10px', 'fontSize': '16px'}
            ),
            width=8
        ),
        dbc.Col(
            dbc.Button("Search", id='search-button', n_clicks=0, color="primary"),
            width=4
        )
    ], className="mb-4"),
    html.Div(id='search-results', className="d-flex flex-wrap"),
    # Hidden div to store selected video URL
    html.Div(id="selected-video", style={"display": "none"}),
    # Modal for Video Player
    dbc.Modal(
        [
            dbc.ModalHeader(dbc.ModalTitle("Video Player")),
            dbc.ModalBody(
                html.Video(id="video-player", controls=True, style={"width": "100%"})
            ),
            dbc.ModalFooter(
                dbc.Button("Close", id="close-modal", n_clicks=0)
            ),
        ],
        id="video-modal",
        is_open=False,
        size="xl",
    )
], fluid=True)

# -------------------------------
# 7. Callbacks
# -------------------------------

# search callback that uses key phrases and topics.
@app.callback(
    Output('search-results', 'children'),
    [Input('search-button', 'n_clicks')],
    [State('search-input', 'value')]
)
def update_results(n_clicks, search_value):
    if not search_value or search_value.strip() == "":
        return html.P("Please enter a keyword to search.")
    
    query = search_value.lower()
    
    # a helper function to compute a simple score.
    # Weight key phrase matches more heavily, then topic matches.
    def compute_score(row):
        score = 0
        # Split the comma-separated key phrases and topics into lists
        key_phrases_list = [phrase.strip() for phrase in row['key_phrases'].lower().split(",") if phrase.strip()]
        topics_list = [phrase.strip() for phrase in row['topics'].lower().split(",") if phrase.strip()]
        
        # Use an exact match check using any() to avoid substring issues.
        if any(token == query for token in key_phrases_list):
            score += 2
        if any(token == query for token in topics_list):
            score += 2
            
        # For the transcript, we still use a regex with word boundaries.
        import re
        transcript_pattern = r'\b' + re.escape(query) + r'\b'
        if re.search(transcript_pattern, row['normalized_text'].lower()):
            score += 1
            
        return score

    # For debugging, print out scores for each video.
    for idx, row in df.iterrows():
        print(f"Video {row['video_title']}: key_phrases={row['key_phrases']} | score={compute_score(row)}")
    
    # Compute scores and add them to the DataFrame.
    df['score'] = df.apply(compute_score, axis=1)
    
    # Only consider videos with a score >= 2 (i.e. term must appear in key phrases or topics).
    filtered_df = df[df['score'] >= 2]
    
    if filtered_df.empty:
        return html.P(f"No videos found for keyword: '{search_value}'")
    
    cards = []
    for idx, row in filtered_df.iterrows():
        card = dbc.Card(
            [
                dbc.CardImg(src=row["thumbnail"], top=True, style={"height": "200px", "object-fit": "cover"}),
                dbc.CardBody(
                    [
                        html.H4(row["video_title"], className="card-title"),
                        html.P("Key Phrases: " + row["key_phrases"], className="card-text"),
                        html.P("Topics: " + row["topics"], className="card-text"),
                        dbc.Button("Play Video", id={"type": "play-button", "index": idx}, n_clicks=0,
                                   color="primary", className="mt-2")
                    ]
                )
            ],
            style={"width": "18rem", "margin": "10px"}
        )
        cards.append(dbc.Col(card, width="auto"))
    
    return dbc.Row(cards)

# Combined callback for opening/closing the modal.
@app.callback(
    [Output("selected-video", "children"),
     Output("video-modal", "is_open")],
    [Input({"type": "play-button", "index": dash.dependencies.ALL}, "n_clicks"),
     Input("close-modal", "n_clicks")],
    [State("video-modal", "is_open")],
    prevent_initial_call=True
)
def toggle_modal(play_nclicks_list, close_click, is_open):
    # Only proceed if at least one play button is clicked or the close button is clicked.
    if not any(n and n > 0 for n in play_nclicks_list) and not close_click:
        return dash.no_update, is_open

    ctx = callback_context
    if not ctx.triggered:
        return dash.no_update, is_open

    triggered_prop = ctx.triggered[0]["prop_id"]
    if triggered_prop == "close-modal.n_clicks":
        return dash.no_update, False
    else:
        import json
        try:
            triggered_id = json.loads(triggered_prop.split(".")[0])
        except Exception as e:
            return dash.no_update, is_open
        idx = triggered_id["index"]
        video_url = f"/videos/{df.loc[idx, 'video_title']}.mp4"
        return video_url, True

# Callback to update the video player's source in the modal.
@app.callback(
    Output("video-player", "src"),
    Input("selected-video", "children")
)
def update_video_src(selected_video):
    if selected_video:
        return selected_video
    return dash.no_update

# -------------------------------
# 8. Run the Server
# -------------------------------
app.run(mode='external', host='0.0.0.0', port=8060, debug=True)
```
